#Age 전처리

```python
def extract_age(text):
  pattern = r'\d+'

  kanji_to_arabic = {
      '一': 1, '二': 2, '三': 3, '四': 4, '五': 5,
      '六': 6, '七': 7, '八': 8, '九': 9, '十': 10
  }

  match = re.search(pattern, text)
  if match:
      return int(match.group())
  else:
      numeric_age = 0
      chars = list(text)
      i = 0
      while i < len(chars):
          char = chars[i]
          if char in kanji_to_arabic:
              if char == '十':
                  if i == 0:
                      numeric_age += 10
                  else:
                      numeric_age += kanji_to_arabic[char]
              else:
                  numeric_age += kanji_to_arabic[char]
          i += 1
      return numeric_age

train['Age'] = train['Age'].apply(lambda age_text: extract_age(age_text) if isinstance(age_text, str) and age_text else None)
test['Age'] = test['Age'].apply(lambda age_text: extract_age(age_text) if isinstance(age_text, str) and age_text else None)
```

#DurationOfPitch 전처리


```python
def convert_duration(item):
    if isinstance(item, str) and item:
        total_seconds = 0
        if '分' in item:
            minutes = int(item.replace('分', ''))
            total_seconds += minutes * 60  # 분을 초로 변환
        if '秒' in item:
            seconds = int(item.replace('秒', ''))
            total_seconds += seconds  # 이미 초이므로 그대로 더함
        return total_seconds
    return pd.NA

train['DurationOfPitch'] = train['DurationOfPitch'].apply(convert_duration)
test['DurationOfPitch'] = test['DurationOfPitch'].apply(convert_duration)
```

```python
# 필요한 라이브러리 가져오기
from rapidfuzz import process, fuzz
import pandas as pd
import unicodedata
import re

# 전처리 함수 정의
def preprocess_text(text):
    text = unicodedata.normalize('NFKC', text)
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

# 퍼지 매칭 함수 정의
def fuzzy_category_mapping(text, choices, scorer=fuzz.WRatio):
    match = process.extractOne(text, choices, scorer=scorer)
    if match:
        return choices[match[2]]

# 카테고리 매핑
category_mapping = ['basic', 'standard', 'deluxe', 'super deluxe', 'king']

# 데이터 전처리 및 카테고리 할당
train['ProductPitched'] = train['ProductPitched'].apply(preprocess_text)
train['ProductPitched'] = train['ProductPitched'].apply(
    fuzzy_category_mapping, choices=category_mapping)

# 데이터 전처리 및 카테고리 할당
test['ProductPitched'] = test['ProductPitched'].apply(preprocess_text)
test['ProductPitched'] = test['ProductPitched'].apply(
    fuzzy_category_mapping, choices=category_mapping)

# 결과 출력
print(train['ProductPitched'])
```

#NumberOfTrips 전처리


```python
# NumberOfTrips 속성 변형 함수
def extract_number_of_trips(trips):
    if isinstance(trips, str):
        match = re.search(r'\d+', trips)  # 숫자 추출
        if match:
            return int(match.group())  # 숫자를 정수로 변환하여 반환
    return trips  # 숫자가 없으면 원래 값 반환

train['NumberOfTrips'] = train['NumberOfTrips'].apply(lambda trips: 2 if trips == '半年に1回' else trips)
train['NumberOfTrips'] = train['NumberOfTrips'].apply(lambda trips: 1 if trips == '四半期に1回' else trips)
train['NumberOfTrips'] = train['NumberOfTrips'].apply(extract_number_of_trips)

test['NumberOfTrips'] = test['NumberOfTrips'].apply(lambda trips: 2 if trips == '半年に1回' else trips)
test['NumberOfTrips'] = test['NumberOfTrips'].apply(lambda trips: 1 if trips == '四半期に1回' else trips)
test['NumberOfTrips'] = test['NumberOfTrips'].apply(extract_number_of_trips)
```

#MonthlyIncome 전처리

```python
def convert_monthly_income(income):

    #문자열에서 '月収'와 '万円' 포함된 데이터 처리
    if isinstance(income, str):
        match = re.search(r'(\d+(\.\d+)?)', income)
        if '月収' in income or '万円' in income:
        # 정규 표현식으로 숫자 추출
          if match:
              # 숫자 변환 및 만엔 단위 변환
              return float(match.group(0)) * 10000  # 만엔을 원으로 변환
        else:
          return float(match.group(0))
    return pd.NA  # 위 조건에 맞지 않는 경우 NaN 반환

train['MonthlyIncome'] = train['MonthlyIncome'].apply(convert_monthly_income)
test['MonthlyIncome'] = test['MonthlyIncome'].apply(convert_monthly_income)
```

#customer_info 전처리
* 구분자로 먼저 차량 보유 여부, 혼인 여부, 아이 동반 여부 컬럼 나누기




```python
import pandas as pd
import re

# 문자열을 구분자로 나누어 새로운 컬럼으로 생성하는 함수 정의
def split_customer_info(info):
    info = re.sub(r'[^\w\s、/]', ' ', info)  # 특수 문자 공백처리
    info = re.sub(r'[\t\n、/]', ' ', info)  # 탭과 줄 바꿈 문자 공백 처리
    info = re.sub(r'\s+', ' ', info)  # 다중 공백을 단일 공백으로 변환

    # 정보 분리
    parts = re.split(' ', info)  # 공백 기호로 분리
    parts = [part.strip() for part in parts]  # 각 부분의 양쪽 공백 제거

    # 기본값 설정
    while len(parts) < 3:  # 필요한 만큼의 컬럼 수에 맞추기
        parts.append('Unknown')

    return pd.Series(parts[:3], index=['Marital_Status', 'Car_Ownership', 'Has_Children'])

# 데이터프레임에 적용
train[['Marital_Status', 'Car_Ownership', 'Has_Children']] = train['customer_info'].apply(split_customer_info)
test[['Marital_Status', 'Car_Ownership', 'Has_Children']] = train['customer_info'].apply(split_customer_info)

# 결과 확인
train[['Marital_Status', 'Car_Ownership', 'Has_Children']]
```

```python
# dtype을 수정하는 함수
def adjust_dtypes(df):
    # MonthlyIncome 컬럼을 float로 변환
    df['MonthlyIncome'] = pd.to_numeric(df['MonthlyIncome'], errors='coerce')
    df['DurationOfPitch'] = pd.to_numeric(df['DurationOfPitch'], errors='coerce')

    # Object 타입 중 범주형 데이터를 category 타입으로 변환
    category_columns = ['TypeofContact', 'Occupation', 'Gender',
                        'ProductPitched', 'Designation',
                        'Marital_Status']

    for col in category_columns:
        df[col] = df[col].astype('category')

    # NumberOfPersonVisiting, NumberOfFollowups, PreferredPropertyStar, NumberOfTrips 컬럼을 float로 변환
    df['NumberOfPersonVisiting'] = df['NumberOfPersonVisiting'].astype(float)
    df['NumberOfFollowups'] = df['NumberOfFollowups'].astype(float)
    df['PreferredPropertyStar'] = df['PreferredPropertyStar'].astype(float)
    df['NumberOfTrips'] = df['NumberOfTrips'].astype(float)
    df[['Car_Ownership', 'Has_Children']] = df[['Car_Ownership', 'Has_Children']].astype(int)

    return df

# dtype을 수정한 데이터프레임
train = adjust_dtypes(train)
test = adjust_dtypes(test)
# 결과 확인
print(train.dtypes)
print(test.dtypes)
```
