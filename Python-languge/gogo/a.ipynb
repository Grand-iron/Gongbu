```python

b['Genre Type'] = b['Fav genre'].apply(lambda x: 
                                       'Energetic' if x in ['Rock', 'EDM', 'Hip hop', 'Pop', 'Rap'] 
                                       else 'Calm' if x in ['Jazz', 'Classical', 'Lofi']
                                       else 'Emotional' if x in ['R&B','Folk','Gospel']
                                       else 'Intense' if x in ['Metal','Country','Video game music']
                                       else 'Mixed')
```

```python
# 음악 장르별 청취량 분포 확인
sns.countplot(x='Genre Type', data=b)
plt.title('Favorite Music Genres')
plt.xticks(rotation=90)
plt.show()
```

```python
# 그룹별 Anxiety 점수의 정규화된 빈도 계산
anxiety_counts = b.groupby('Genre Type')['Anxiety'].value_counts(normalize=True)

# 시각화
anxiety_counts.unstack().plot(kind='bar')
plt.title('Normalized Frequency of Anxiety Scores by Music Genre Type')
plt.xlabel('Genre Type')
plt.ylabel('Normalized Frequency')
plt.show()
```

```python


# 그래프를 한 번에 출력하기 위해 subplots 사용
fig, axes = plt.subplots(2, 2, figsize=(16, 12))  # 2x2 그래프 레이아웃

# 각 감정 별로 그래프 생성
emotions = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
for i, emotion in enumerate(emotions):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    emotion_counts = a.groupby('Music effects')[emotion].value_counts(normalize=True)
    emotion_counts.unstack().plot(kind='bar', ax=ax)
    
    ax.set_yticklabels(['{:.0f}%'.format(x*100) for x in ax.get_yticks()])
    ax.set_title(f'음악효능의 유무와 {emotion}의 관계')
    ax.set_xlabel('음악효능')
    ax.set_ylabel('평균 점수')
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
anxiety_counts = a.groupby('Music effects')['Anxiety'].value_counts(normalize=True)

# 시각화
fig, ax = plt.subplots(figsize=(12, 8))  # 그래프 크기 조절
anxiety_counts.unstack().plot(kind='bar', ax=ax)
ax.set_yticklabels(['{:.0f}%'.format(x*100) for x in ax.get_yticks()])
plt.title('상세 점수 관계')
plt.xlabel('음악효능')
plt.ylabel('점수 비율')
plt.xticks(rotation=90)  # x축 라벨 회전
plt.tight_layout()
plt.show()
```
